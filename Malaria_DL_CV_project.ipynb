```python
# Importing the required libraries to create the models:

from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten, BatchNormalization, Dropout
```

### 4.1 First Model architecture

```python
# We will build our first model using 3x3 kernels with 3 convolution layers:

# Start Neural Network

model1 = Sequential()

# 1st convolution layer:
model1.add(Conv2D(filters = 16, kernel_size = (3,3), padding = 'same', activation = 'relu', input_shape = (100,100,3)))
model1.add(MaxPool2D()) # default size

# 2nd convolution layer:
model1.add(Conv2D(filters = 32, kernel_size = (3,3), padding = 'same', activation = 'relu'))
model1.add(MaxPool2D()) # default size

# 3rd convolution layer:
model1.add(Conv2D(filters = 64, kernel_size = (3,3), padding = 'same', activation = 'relu'))
model1.add(MaxPool2D()) # default size
model1.add(Dropout(0.25))

#Fully connected 1st layer
model1.add(Flatten()) 
model1.add(Dense(512,activation = 'relu')) 
model1.add(Dropout(0.25))

#Fully connected final layer
model1.add(Dense(2, activation = 'softmax')) 

# Model Summary:
print()
print('Model Summary: ')
model1.summary()
print()
print('--------------------------------------------')
print()

# Compiling the model:
model1.compile(loss = 'categorical_crossentropy', optimzier = 'Adam', metrics = ['accuracy'])

# Fitting the model to our data:
history = model1.fit(X_train,y_train, batch_size = 100, epochs = 20, validation_data = (X_test, y_test))

# Evaluating on the test data:
score = model1.evaluate(X_test, y_test, verbose=0)


# Printing loss and accuracy on test data:
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

### 4.3 Model Architecture using Hyperparameter tuning:

### We have created the above models by specifying the number of filters, filter size etc. but, like this can create 'n' number of models which will consume a lot of time.

### In order to avoid this we will be using hyperparameter tuning using the 'kerastuner' library to select the best values for number of filters, filters sizes etc.

```python
# Importing the requiredlibraries:

!pip install keras-tuner
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from kerastuner import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameters
```

```python
# Checking the tensorflow version:
print(tf.__version__)
```

```python
# Creating a function to build our model:

def create_model(hp):  
  
  # Starting the neural network
  model = keras.Sequential([
    
    # First Convolution layer
    keras.layers.Conv2D(
        
        filters=hp.Int('filter_1', min_value=16, max_value=128, step=16),
        kernel_size= hp.Choice('kernel_1', values = [3,5]),
        activation='relu',
        input_shape= ()

    ),

    # Second Convolution layer
    keras.layers.Conv2D(
        
        filters=hp.Int('filter_2', min_value=32, max_value=128, step=16),
        kernel_size=hp.Choice('kernel_2', values = [3,5]),
        activation='relu'

    ),


    # Third Convolution layer
    keras.layers.Conv2D(
        
        filters=hp.Int('filter_3', min_value=64, max_value=128, step=16),
        kernel_size=hp.Choice('kernel_3', values = [3,5]),
        activation='relu'

    ),

    keras.layers.Flatten(),

    keras.layers.Dense(
        units=hp.Int('dense_1_units', min_value=32, max_value=128, step=16),
        activation='relu'
    ),

    keras.layers.Dense(2, activation='softmax')
  ])
  
  #Compiling our model:
  model.compile(optimizer= keras.optimizers.Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3])),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
  
  return model
```

```python
# Setting the tuner parameters:

tuner_params = RandomSearch(build_model,
                            objective='val_accuracy',
                            max_trials=5,directory='output',project_name="Mnist Fashion")
```

```python
# Performing the parameter searching using ourtuner:

tuner_params.search(train_images,train_labels,epochs=5,validation_split=0.25)
```

```python
# Extracting the best performing model:

best_model = tuner_params.get_best_models(num_models=1)[0]
```

```python
# Extracting the summary ofour best model:
best_model.summary()
```

```python
# Creating our final model using the best model parameters:

best_model.fit(train_images, train_labels, epochs=10, validation_split=0.1, initial_epoch=3)
```
